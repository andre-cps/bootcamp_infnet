4) Refaça o exercício usando o conceito de Classe e Objeto. A sua Classe deve ser aplicada a qualquer contexto, não somente ao problema acima. De forma que ela seja genérica. Cada método deve representar uma única ação do Objeto e não, especificamente, uma função que executa diversas ações. Pensem em ações atômicas. Boa sorte !

```python
class WordCounter:
    def __init__(self, arquivo):
        self.arquivo = arquivo
        self.palavras = []
        self.numero_de_palavras = 0
        self.checking = 0
        self.dicionario = dict()

    def carregar_arquivo(self):
        with open(self.arquivo, 'r') as texto:
            texto = texto.read().strip().lower()
            self.palavras = texto.split(" ")

    def contar_palavras(self):
        for i in palavras:
            self.numero_de_palavras += 1
    
    def criar_dicionario(self):
        for palavra in palavras:
            if palavra in self.dicionario:
                self.dicionario[palavra] += 1 
            else:
                self.dicionario[palavra] = 1
        
        self.dicionario = dict(sorted(self.dicionario.items(), key=lambda x: x[1], reverse=True))

    def calcular_percentual(self):
        for i in self.dicionario:
            self.dicionario[i] = round((self.dicionario[i] / self.numero_de_palavras) * 100, 3)

    def checagem(self):
        for i in self.dicionario:
            self.checking += self.dicionario[i]
    
    def imprimir_resultados(self):
        print(self.checking)
        print(self.dicionario)
```

```python
WordCounter = WordCounter('C:/Users/andre/Documents/GitHub/Bootcamp_TurmaNoite/Conteudo_02/Alice.txt')

WordCounter.carregar_arquivo()
WordCounter.contar_palavras()
WordCounter.criar_dicionario()
WordCounter.calcular_percentual()
WordCounter.checagem()
WordCounter.imprimir_resultados()
```

5) Repare que palavras como 'it' e 'it,' serão contabilizados como palavras diferentes por causa da pontuação. Pense em uma forma de se remover possíveis pontuações. Outro caso é 'The' e 'the', que também será processado como palavras diferentes, como resolver isso ? Observe outras situações similares que precisariam de algum pré-processamento.

 Outro ponto a se observar é que a quantidade de palavras que são artigos, preposições, por exemplos, são muito comuns em textos. Se queremos fazer um análise de relevância de palavras presentes em textos,devemos ou não considerar estes tipos de palavras ?  

```python
#3 - B

with open('C:/Users/andre/Documents/GitHub/Bootcamp_TurmaNoite/Conteudo_02/Alice.txt', 'r') as arquivo:
    texto = arquivo.read()

    #Arrumando o texto
    texto = texto.strip().lower().replace('\n', ' ')

    #Remover pontuações
    texto_sem_pontuacao = ""
    pontuacoes = '''!()-[]{};:'"\,<>./?@#$%^&*_~ï»¿'''

    for i in texto:
        if i not in pontuacoes:
            texto_sem_pontuacao += i

    #Separando palavras
    palavras_nao_confirmadas = []
    palavras_nao_confirmadas = texto_sem_pontuacao.split(" ")

    #Verificando se é uma palavra
    palavras = []
    for i in palavras_nao_confirmadas:
        if i.isalpha() == True:
            palavras.append(i)
    
    #Criando o dicionario com as palavras
    d = dict()
    #Para cada "Palavra" em "palavras":
    for palavra in palavras:
        #Se palavra já existe no dicionario:
        if palavra in d:
            #Incremente mais 1 a soma
            d[palavra] = d[palavra] + 1
            #Caso não existe adicione apenas o valor 1
        else:
            d[palavra] = 1

    #Classificando o dicionario do maior para o menor
    dicinario_classificado = dict(sorted(d.items(), key=lambda x: x[1], reverse=True))

    #Passando preposicoes e artigos para outro dicionario
    d_prepo_art = dict()
    lista_prepo_arti = ["of", "in", "to", "for", "with", "on", "at", "by", "from", "about", "into", "through", "over", "after", "under", "before", "between", "among", "the", "a", "an"]

    for i in dicinario_classificado:
        if i in lista_prepo_arti:
            d_prepo_art[i] = dicinario_classificado[i]

    #Retirando as preposicoes e artigos do dicionario de palavras
    for chave in lista_prepo_arti:
        if chave in dicinario_classificado:
            del dicinario_classificado[chave]

    #Criando o percentual de cada palavra do dicionario de palavras
    numero_palavras = sum(dicinario_classificado.values())

    for i in dicinario_classificado:
        dicinario_classificado[i] = (dicinario_classificado[i] / numero_palavras) * 100

    #Criando o percentual de cada palavra do dicionario de preposicoes e artigos
    numero_pre_art = sum(d_prepo_art.values())

    for i in d_prepo_art:
        d_prepo_art[i] = round((d_prepo_art[i] / numero_pre_art) * 100, 2)

    #Somando a divisão para ver se chega a 100
    checking_palavras = round(sum(dicinario_classificado.values()))
    checking_prep_art = round(sum(d_prepo_art.values()))


print(f'''Foi criado dois dicionarios com base no texto Alice s Adventures in Wonderland, um dicionario de palavras e outro de preposições e artigos.
No dicionario de palavras, há um total de {numero_palavras} palavras.''')

print()

print(f'''Verificação: As palavras estão com seus percentual correto? 

Total calculado = {checking_palavras}

Se o total calculado acima for igual a 100 ou 99, tudo em ordem.''')

print()

print(dicinario_classificado)

print()

print(f'''No dicionario de preposições e artigos, há um total de {numero_pre_art} preposições e artigos.''')

print()

print(f'''Verificação: As preposições e artigos estão com seus percentual correto? 

Total calculado = {checking_prep_art}

Se o total calculado acima for igual a 100 ou 99, tudo em ordem.''')

print()

print(d_prepo_art)
```
